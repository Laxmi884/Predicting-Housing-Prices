# Finding Categorical Values and converting it to numercial format for machine learning models

```python
# finding categorical values
cat_var = []
for i in list(df.columns):
  if df[i].dtype == 'object':
    cat_var.append(i)
cat_var
```

```python
from numpy.ma.core import append
#finding unique values in each categorical values
list_unique = []
for i in cat_var:
  list_unique.append(list(df[i].unique()))
ab = dict(zip(cat_var,list_unique) )
ab['House_Style']
```

```python
def convert(x,col_name='abab'):
  a=ab[col_name]
  b= a.index(x)
  return b

for i in cat_var:
  df[i+'vec'] = df[i].apply(convert,col_name=i)

df.head()


```
